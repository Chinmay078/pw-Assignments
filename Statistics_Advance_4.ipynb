Q1: Difference between a t-test and a z-test

t-test: Used when the population standard deviation is unknown, and the sample size is small (<30).
Example: Testing whether the mean height of a small group differs from the population mean.
z-test: Used when the population standard deviation is known, or the sample size is large (≥30).
Example: Testing the mean score of a large class against a known population mean.


Q2: One-tailed vs. Two-tailed Tests

One-tailed test: Tests for an effect in a specific direction.
Example: Checking if a drug increases blood pressure.
Two-tailed test: Tests for an effect in either direction.
Example: Testing if a drug changes blood pressure (increase or decrease).


Q3: Type 1 and Type 2 Errors

Type 1 error: Rejecting a true null hypothesis (false positive).
Example: Convicting an innocent person in court.
Type 2 error: Failing to reject a false null hypothesis (false negative).
Example: Letting a guilty person go free.

Q8: Margin of Error (MOE)

Definition: The maximum expected difference between the sample statistic and population parameter.
Effect of sample size: Larger samples reduce MOE.
Example: Estimating voter preferences with a large sample reduces error.
