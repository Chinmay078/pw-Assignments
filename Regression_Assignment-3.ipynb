## Q1. What is Ridge Regression, and how does it differ from ordinary least squares regression?

Ridge Regression is a regularized linear regression technique that adds an L2 penalty (sum of squared coefficients) to the loss function.

**Difference from Ordinary Least Squares (OLS):**
- OLS minimizes only the residual sum of squares.
- Ridge minimizes the residual sum of squares plus an L2 penalty.
- Ridge reduces overfitting and handles multicollinearity better.
- OLS may produce unstable coefficients when predictors are highly correlated.

## Q2. What are the assumptions of Ridge Regression?

The assumptions are:
- Linear relationship between predictors and target.
- Independence of observations.
- Homoscedasticity (constant variance of errors).
- Normally distributed residuals.
- Little or no perfect multicollinearity.
- Features should be standardized before applying Ridge Regression.

## Q3. How do you select the value of the tuning parameter (lambda) in Ridge Regression?

The regularization parameter (lambda or alpha) is usually selected using:
- Cross-validation (e.g., RidgeCV)
- Grid Search
- Validation datasets

The value giving the lowest prediction error is chosen.

## Q4. Can Ridge Regression be used for feature selection? If yes, how?

Ridge Regression does not perform true feature selection because it shrinks coefficients toward zero but rarely makes them exactly zero. It is mainly used to reduce the effect of less important features rather than remove them.

## Q5. How does the Ridge Regression model perform in the presence of multicollinearity?

Ridge Regression performs well because it shrinks correlated coefficients, reducing variance and improving model stability without removing variables.

## Q6. Can Ridge Regression handle both categorical and continuous independent variables?

Yes. Continuous variables can be used directly. Categorical variables must first be converted into numerical form using techniques such as one-hot encoding before fitting the model.

## Q7. How do you interpret the coefficients of Ridge Regression?

The coefficients represent the effect of each predictor on the target variable, but they are shrunk toward zero because of regularization. Larger absolute coefficients indicate a stronger influence on the prediction.

## Q8. Can Ridge Regression be used for time-series data analysis? If yes, how?

Yes. Ridge Regression can be applied to time-series forecasting by using lagged variables or engineered time-based features as predictors. Regularization helps reduce overfitting when many correlated lag features are included.
