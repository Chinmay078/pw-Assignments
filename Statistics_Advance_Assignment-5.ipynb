## Q1. 95% Confidence Interval

```
from scipy.stats import norm
mean,sd,n=50,8,100
z=norm.ppf(0.975)
margin=z*sd/(n**0.5)
print((mean-margin,mean+margin))
```

## Q2. Chi-square Goodness of Fit

```
from scipy.stats import chisquare
observed=[25,20,22,18,15]
expected=[20,20,20,20,20]
print(chisquare(observed,f_exp=expected))
```

## Q3. Chi-square Test of Independence

```
import pandas as pd
from scipy.stats import chi2_contingency
table=pd.DataFrame([[20,15],[10,25],[15,20]])
print(chi2_contingency(table))
```

## Q4. CI for Proportion

```
from statsmodels.stats.proportion import proportion_confint
print(proportion_confint(60,500,alpha=0.05,method='normal'))
```

## Q5. 90% Confidence Interval

```
from scipy.stats import norm
mean,sd,n=75,12,100
z=norm.ppf(0.95)
m=z*sd/(n**0.5)
print((mean-m,mean+m))
```

## Q6. Plot Chi-square Distribution

```
import numpy as np, matplotlib.pyplot as plt
from scipy.stats import chi2
x=np.linspace(0,30,300)
plt.plot(x,chi2.pdf(x,10))
plt.xlabel("Chi-square"); plt.ylabel("Density")
plt.show()
```

## Q7. 99% CI for Coke Preference

```
from statsmodels.stats.proportion import proportion_confint
print(proportion_confint(520,1000,alpha=0.01,method='normal'))
```

## Q8. Chi-square Goodness of Fit (Die)

```
from scipy.stats import chisquare
obs=[15,20,18,17,15,15]
exp=[100/6]*6
print(chisquare(obs,f_exp=exp))
```

## Q11. One-Sample t-Test

```
import numpy as np
from scipy.stats import ttest_1samp
np.random.seed(1)
sample=np.random.normal(72,10,30)
print(ttest_1samp(sample,70))
```
