## Q1. Difference between Linear Regression and Logistic Regression

Linear Regression predicts continuous values, while Logistic Regression predicts categorical outcomes using the sigmoid function.

Example: Predicting whether an email is spam (Logistic Regression).

## Q2. Cost function in Logistic Regression

Logistic Regression uses **Log Loss (Binary Cross-Entropy)**. It is optimized using Gradient Descent or similar optimization algorithms.

## Q3. Regularization in Logistic Regression

Regularization prevents overfitting.
- L1 (Lasso): Feature selection.
- L2 (Ridge): Shrinks coefficients.

## Q4. ROC Curve

The ROC curve plots the True Positive Rate against the False Positive Rate. The Area Under the Curve (AUC) measures classifier performance.

## Q5. Feature selection techniques

Methods include Recursive Feature Elimination (RFE), L1 regularization, correlation analysis, mutual information, and tree-based feature importance.

## Q6. Handling imbalanced datasets

Use oversampling (SMOTE), undersampling, class weights, threshold tuning, and metrics such as Precision, Recall, F1-score, and ROC-AUC.

## Q7. Common issues and challenges

Common issues include multicollinearity, missing values, class imbalance, overfitting, and outliers. Solutions include feature selection, regularization, imputation, balancing the dataset, and scaling features.
