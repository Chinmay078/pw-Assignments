## Q1. Assumptions of ANOVA

- Independence of observations.
- Normally distributed residuals.
- Homogeneity of variances.

Violations may lead to unreliable F-tests and incorrect conclusions.

## Q2. Types of ANOVA

- One-Way ANOVA: One independent variable.
- Two-Way ANOVA: Two independent variables and interaction.
- Repeated Measures ANOVA: Same subjects measured multiple times.

## Q3. Partitioning of Variance in ANOVA

ANOVA divides the Total Sum of Squares (SST) into:
- Explained Sum of Squares (SSA/SSB)
- Residual/Error Sum of Squares (SSE)

SST = SSA + SSE.

## Q4. One-Way ANOVA in Python

```
import pandas as pd
from scipy.stats import f_oneway

group1=[10,12,11,13,12]
group2=[15,16,14,17,15]
group3=[20,19,18,21,20]

f,p=f_oneway(group1,group2,group3)
print("F-statistic:",f)
print("p-value:",p)
```

## Q5. Two-Way ANOVA in Python

```
import pandas as pd
from statsmodels.formula.api import ols
import statsmodels.api as sm

df=pd.DataFrame({
'Score':[10,12,11,15,14,16,18,20,19],
'Program':['A','A','A','B','B','B','C','C','C'],
'Experience':['Novice','Experienced','Novice']*3
})

model=ols('Score ~ C(Program)+C(Experience)+C(Program):C(Experience)',data=df).fit()
print(sm.stats.anova_lm(model,typ=2))
```

## Q6. Interpretation

If F = 5.23 and p = 0.02 (<0.05), reject the null hypothesis. There is a statistically significant difference between at least one pair of group means.

## Q7. Missing Data in Repeated Measures ANOVA

Common approaches:
- Listwise deletion
- Mean/median imputation
- Multiple imputation
- Mixed-effects models (preferred)

Different methods may affect statistical power and bias.

## Q8. Common Post-hoc Tests

- Tukey HSD
- Bonferroni
- Scheffé
- Dunnett

Post-hoc tests are used after a significant ANOVA result to identify which groups differ.

## Q11. Two-Sample t-test

```
from scipy.stats import ttest_ind

control=[65,68,70,66,69]
experiment=[75,78,74,77,76]

print(ttest_ind(control,experiment))
```

## Q12. Repeated Measures ANOVA

```
import pandas as pd
from statsmodels.stats.anova import AnovaRM

df=pd.DataFrame({
'Subject':[1,1,1,2,2,2,3,3,3],
'Store':['A','B','C']*3,
'Sales':[100,105,98,110,112,108,95,97,96]
})

print(AnovaRM(df,'Sales','Subject',within=['Store']).fit())
```
