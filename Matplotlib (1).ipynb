Q1: What is Matplotlib? Why is it used?
Matplotlib is a popular data visualization library in Python. It is used to create static, interactive, and animated visualizations in Python. It provides flexibility in designing and customizing plots.

The Pyplot module of Matplotlib is used to create various visualizations such as line plots, bar plots, scatter plots, and more.

Five types of plots in Pyplot:

Line plot
Scatter plot
Bar plot
Histogram
Box plot

Q2: What is a scatter plot?
A scatter plot is used to represent the relationship between two variables by displaying data points on a 2D plane. Each point represents one observation.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y, color='blue', alpha=0.7)
plt.title("Scatter Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

Q3: Why is the subplot() function used?
The subplot() function is used to create multiple plots in a single figure. It divides the plotting area into a grid, allowing multiple plots to be visualized in different grid sections.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
x = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Subplots
plt.subplot(2, 2, 1)
plt.plot(x, y1, color='blue', label="Line 1")
plt.title("Line 1")
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(x, y2, color='green', label="Line 2")
plt.title("Line 2")
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(x, y3, color='red', label="Line 3")
plt.title("Line 3")
plt.legend()

plt.subplot(2, 2, 4)
plt.plot(x, y4, color='orange', label="Line 4")
plt.title("Line 4")
plt.legend()

plt.tight_layout()
plt.show()
```

Q4: What is a bar plot? Why is it used?
A bar plot is used to represent categorical data with rectangular bars. The height (or length) of each bar corresponds to the value of the category. A horizontal bar plot is the same but with horizontal bars.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Bar plot
plt.bar(company, profit, color='purple', alpha=0.7)
plt.title("Company Profits (Bar Plot)")
plt.xlabel("Company")
plt.ylabel("Profit")
plt.show()

# Horizontal bar plot
plt.barh(company, profit, color='teal', alpha=0.7)
plt.title("Company Profits (Horizontal Bar Plot)")
plt.xlabel("Profit")
plt.ylabel("Company")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Bar plot
plt.bar(company, profit, color='purple', alpha=0.7)
plt.title("Company Profits (Bar Plot)")
plt.xlabel("Company")
plt.ylabel("Profit")
plt.show()

# Horizontal bar plot
plt.barh(company, profit, color='teal', alpha=0.7)
plt.title("Company Profits (Horizontal Bar Plot)")
plt.xlabel("Profit")
plt.ylabel("Company")
plt.show()
```

Q5: What is a box plot? Why is it used?


A box plot (or box-and-whisker plot) is used to represent the distribution of data. It shows the median, quartiles, and potential outliers in a dataset, helping to understand the spread and skewness of the data.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Box plot
plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'], patch_artist=True)
plt.title("Box Plot")
plt.ylabel("Values")
plt.show()
```
