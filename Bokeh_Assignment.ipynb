## Q1. How can you create a Bokeh plot using Python code?

```python
from bokeh.plotting import figure, show

x = [1, 2, 3, 4, 5]
y = [2, 5, 8, 2, 7]

p = figure(title="Simple Bokeh Plot", x_axis_label="X", y_axis_label="Y")
p.line(x, y, line_width=2)
show(p)
```

## Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

Glyphs are the visual elements used to represent data in a Bokeh plot, such as **line, circle, square, triangle, vbar, and scatter**.

Example:


```python
from bokeh.plotting import figure, show

p = figure(title="Glyph Example")
p.circle([1,2,3,4],[4,7,2,6], size=10, color="blue")
show(p)
```

## Q3. How can you customize the appearance of a Bokeh plot?

You can customize:
- Plot title
- Axis labels
- Line color and width
- Background color
- Legend position
- Grid lines
- Fonts and themes


```python
from bokeh.plotting import figure, show

p = figure(title="Customized Plot", x_axis_label="X", y_axis_label="Y")
p.line([1,2,3],[2,4,6], legend_label="Line", line_width=3, color="green")
p.legend.location = "top_left"
show(p)
```

## Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in real time?

A **Bokeh Server** allows Python code to run continuously on the server and update plots dynamically based on user interaction or streaming data.

Typical uses:
- Real-time dashboards

## Q5. How can you embed a Bokeh plot into a web page or dashboard using Flask or Django?

A Bokeh plot can be embedded using the `components()` function.

Steps:
1. Create the Bokeh plot.
2. Generate JavaScript and HTML components.
3. Pass them to a Flask or Django template.
4. Render the template to display the interactive plot.


```python
from bokeh.embed import components
from bokeh.plotting import figure

p = figure(title="Embedded Plot")
p.line([1,2,3],[3,2,5])

script, div = components(p)

print(script)
print(div)
```
