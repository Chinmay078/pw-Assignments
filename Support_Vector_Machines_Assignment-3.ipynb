# Q1. Regression metric for predicting house prices

When building an SVM Regression (SVR) model for house price prediction, **RMSE (Root Mean Squared Error)** is generally the best evaluation metric. RMSE measures the average prediction error in the same units as house prices, making it easy to interpret. It also penalizes large errors more heavily than MAE, which is useful because large pricing mistakes can be costly. MSE is also useful during model optimization, while R² is commonly reported to show how well the model explains the variance in the data.

# Q2. MSE or R-squared?

If the goal is to predict house prices as accurately as possible, **MSE (or RMSE)** should be preferred because these metrics directly measure prediction error. R² indicates how well the model explains variation in the target variable but does not indicate the average prediction error. Therefore, use RMSE as the primary metric and R² as a supporting metric.

# Q3. Dataset with many outliers

For datasets containing many outliers, **MAE (Mean Absolute Error)** is usually the best choice because it is less sensitive to extreme values than MSE or RMSE. MAE gives equal importance to every prediction error, making it more robust when outliers are present.

# Q4. MSE and RMSE are very close

When MSE and RMSE indicate similar performance, **RMSE** is generally preferred because it is expressed in the same units as the target variable. This makes it easier to understand and explain to stakeholders. RMSE also clearly represents the average prediction error.

# Q5. Metric for explaining variance

If the objective is to compare different SVR kernels (Linear, Polynomial, and RBF) based on how well they explain the variation in the target variable, the **R² Score (Coefficient of Determination)** is the most appropriate metric. A higher R² indicates that the model explains more of the variance in house prices.
