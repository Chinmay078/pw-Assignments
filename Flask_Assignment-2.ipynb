## Q1. Explain GET and POST methods.

**GET**
- Retrieves data from the server.
- Data is sent in the URL.
- Used for fetching information.
- Less secure for sensitive data.

**POST**
- Sends data to the server.
- Data is sent in the request body.
- Used for form submission and creating resources.
- More secure than GET for sensitive data.


## Q2. Why is `request` used in Flask?

The `request` object is used to access data sent by the client.

It is commonly used to:
- Read form data (`request.form`)
- Read query parameters (`request.args`)
- Read JSON data (`request.get_json()`)
- Check HTTP request method (`request.method`)


## Q3. Why is `redirect()` used in Flask?

The `redirect()` function sends the user from one URL to another.

**Uses:**
- Redirect after successful login
- Redirect after form submission
- Navigate users between pages
- Avoid duplicate form submissions


## Q4. What are templates in Flask? Why is `render_template()` used?

Templates are HTML files stored inside the **templates** folder.

The `render_template()` function is used to:
- Render HTML pages
- Pass data from Flask to HTML
- Create dynamic web pages


## Q5. Create a simple API. Use Postman to test it.

```
from flask import Flask, jsonify

app = Flask(__name__)

@app.route('/api', methods=['GET'])
def api():
    return jsonify({
        "message": "Hello from Flask API",
        "status": "success"
    })

if __name__ == '__main__':
    app.run(debug=True)
```

**Testing in Postman**
1. Run the Flask application.
2. Open Postman.
3. Send a GET request to:
   `http://127.0.0.1:5000/api`
4. Verify the JSON response.
