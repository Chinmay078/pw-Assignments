## Q1. What is an API? Give an example where an API is used in real life.

An API (Application Programming Interface) is a set of rules that allows different software applications to communicate with each other.

Example: A weather app uses a Weather API to fetch live weather information.

## Q2. Give advantages and disadvantages of using API.

**Advantages**
- Easy integration
- Saves development time
- Reusable services
- Automation
- Scalable

**Disadvantages**
- Security risks
- Depends on third-party services
- Rate limits
- Maintenance issues

## Q3. What is a Web API? Differentiate between API and Web API.

| API | Web API |
|---|---|
| Can work locally or remotely | Works over HTTP/HTTPS |
| May use different protocols | Uses web protocols |
| Internet not always required | Internet/network required |
| Example: Java API | Example: REST API |

## Q4. Explain REST and SOAP Architecture. Mention shortcomings of SOAP.

**REST** is a lightweight architectural style that uses HTTP methods and commonly exchanges JSON.

**SOAP** is a protocol based on XML with strong security features.

**SOAP Shortcomings:**
- Complex
- XML only
- Slower
- Higher bandwidth usage

## Q5. Differentiate between REST and SOAP.

| REST | SOAP |
|---|---|
| Architectural style | Protocol |
| JSON/XML | XML only |
| Lightweight | Heavyweight |
| Faster | Slower |
| Easy to develop | More complex |
| Best for web/mobile apps | Best for enterprise applications |
