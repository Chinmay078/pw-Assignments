## Q1. What is the KNN algorithm?

**Answer:**

K-Nearest Neighbors (KNN) is a supervised machine learning algorithm used for both classification and regression. It predicts the output of a new data point by finding the **K nearest neighbors** using a distance metric.

**Working:**
1. Choose K.
2. Compute distances.
3. Select K nearest neighbors.
4. Classification: majority voting.
5. Regression: average of neighbors.

**Advantages**
- Simple and easy to implement.
- No training phase.
- Works well on small datasets.

**Disadvantages**
- Slow on large datasets.
- Sensitive to noise and irrelevant features.
- Requires feature scaling.


## Q2. How do you choose the value of K in KNN?

**Answer:**

- Small K may cause overfitting.
- Large K may cause underfitting.
- Select K using cross-validation.
- A common rule is **K ≈ √N**, where N is the number of training samples.
- Odd K values are preferred for binary classification.


## Q3. What is the difference between KNN classifier and KNN regressor?

**Answer:**

| KNN Classifier | KNN Regressor |
|---|---|
| Predicts classes | Predicts continuous values |
| Uses majority voting | Uses average of neighbors |
| Used for classification | Used for regression |

## Q5. What is the curse of dimensionality in KNN?

**Answer:**

As the number of features increases:
- Distances become less meaningful.
- Accuracy decreases.
- Computation increases.

**Solutions**
- Feature selection
- PCA
- Remove irrelevant features
- Collect more data


## Q6. How do you handle missing values in KNN?

**Answer:**

- Remove missing rows.
- Mean/Median/Mode imputation.
- KNN Imputer.
- Interpolation techniques.

KNN Imputer generally provides better estimates by using neighboring samples.


## Q7. Compare and contrast the performance of the KNN classifier and regressor. Which one is better for which type of problem?

**Answer:**

- **KNN Classifier:** Best for categorical outputs (spam detection, disease classification).
- **KNN Regressor:** Best for numerical outputs (house prices, temperature prediction).

The choice depends on the type of prediction required.

## Q9. What is the difference between Euclidean distance and Manhattan distance in KNN?

**Answer:**

**Euclidean Distance**

\[
d=\sqrt{\sum_{i=1}^{n}(x_i-y_i)^2}
\]

**Manhattan Distance**

\[
d=\sum_{i=1}^{n}|x_i-y_i|
\]

Euclidean measures straight-line distance, while Manhattan measures grid-based distance.


## Q10. What is the role of feature scaling in KNN?

**Answer:**

KNN is a distance-based algorithm. Features with larger values dominate the distance calculation if data is not scaled.

**Common methods**
- Min-Max Scaling
- Standardization (Z-score)

Feature scaling improves the accuracy and fairness of distance calculations.
