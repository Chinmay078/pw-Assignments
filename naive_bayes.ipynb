# Q1. Probability of an employee being a smoker given that he/she uses the health insurance plan

Let:

- P(H) = Probability that an employee uses the health insurance plan = **0.70**
- P(S ∩ H) = Probability that an employee both uses the health insurance plan and is a smoker = **0.40**

Using the conditional probability formula:

\[
P(S|H)=\frac{P(S\cap H)}{P(H)}
\]

Substituting the values:

\[
P(S|H)=\frac{0.40}{0.70}=0.5714
\]

**Answer:** The probability that an employee is a smoker given that they use the health insurance plan is **0.5714 (57.14%)**.

**Interpretation:** Among employees who use the company's health insurance plan, about **57%** are smokers.

# Q2. Difference between Bernoulli Naive Bayes and Multinomial Naive Bayes

| Feature | Bernoulli NB | Multinomial NB |
|---|---|---|
| Input data | Binary (0/1) | Count/Frequency |
| Considers absence of feature | Yes | No |
| Best for | Presence/absence of words | Word counts |
| Common application | Binary text features | Spam filtering, document classification |

**Bernoulli Naive Bayes** works with binary features indicating whether a feature is present or absent. It is useful when the existence of a feature is more important than its frequency.

**Multinomial Naive Bayes** uses the frequency of features. It is widely used in text classification where word counts carry important information.

# Q3. How does Bernoulli Naive Bayes handle missing values?

Bernoulli Naive Bayes does **not** handle missing values directly. Missing values should be processed before training.

Common approaches include:
- Mean or median imputation
- Mode imputation for categorical data
- Removing rows or columns with excessive missing values
- Using pipelines with `SimpleImputer`

Proper preprocessing improves model performance and stability.

# Q4. Can Gaussian Naive Bayes be used for multi-class classification?

Yes. Gaussian Naive Bayes naturally supports **multi-class classification**.

It assumes each feature follows a Gaussian (Normal) distribution within every class. During prediction it computes the posterior probability for every class using Bayes' theorem and predicts the class having the highest probability.

Applications include:
- Iris flower classification
- Medical diagnosis
- Handwritten digit recognition
- Sensor data classification


# Q5. Assignment – Implement and Compare Naive Bayes Models

```
import pandas as pd
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.naive_bayes import BernoulliNB, MultinomialNB, GaussianNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.preprocessing import MinMaxScaler

# Load the Spambase dataset
df = pd.read_csv("spambase.csv")

X = df.iloc[:, :-1]
y = df.iloc[:, -1]

# Scale features for MultinomialNB
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

models = {
    "BernoulliNB": BernoulliNB(),
    "MultinomialNB": MultinomialNB(),
    "GaussianNB": GaussianNB()
}

for name, model in models.items():
    model.fit(X_train, y_train)
    pred = model.predict(X_test)

    print("\n", "="*40)
    print(name)
    print("="*40)
    print("Accuracy :", accuracy_score(y_test, pred))
    print("Precision:", precision_score(y_test, pred))
    print("Recall   :", recall_score(y_test, pred))
    print("F1 Score :", f1_score(y_test, pred))

    cv = cross_val_score(model, X_scaled, y, cv=10)
    print("10-Fold CV Accuracy:", cv.mean())
```

## Discussion

After evaluating the three models, compare their Accuracy, Precision, Recall, and F1-score.

- **BernoulliNB** generally performs well for binary features.
- **MultinomialNB** usually performs better for count or frequency-based text features.
- **GaussianNB** works best when numerical features approximately follow a normal distribution.

The best model is the one that achieves the highest performance on the chosen evaluation metric while maintaining good generalization.

**Limitations**
- Strong independence assumption.
- Performance decreases when features are highly correlated.
- Sensitive to inappropriate feature distributions.
