## Q1. What is MongoDB? Explain non-relational databases in short. In which scenarios it is preferred to use MongoDB over SQL databases?

### Answer:

**MongoDB** is an open-source, document-oriented NoSQL database that stores data in flexible, JSON-like documents (called BSON — Binary JSON). It was developed by MongoDB Inc. and is designed for scalability, high availability, and high performance.

---

**Non-Relational Databases (NoSQL):**  
Non-relational databases do **not** use the traditional table-based relational model. Instead, they store data in formats like:
- **Documents** (MongoDB)
- **Key-Value pairs** (Redis)
- **Column families** (Cassandra)
- **Graphs** (Neo4j)

They are schema-less, meaning each record can have a different structure. They are horizontally scalable and suitable for large volumes of unstructured or semi-structured data.

---

**When to prefer MongoDB over SQL:**

| Scenario | Reason |
|---|---|
| Unstructured or variable data | No fixed schema needed |
| Rapid application development | Easy to change data model |
| High-volume read/write operations | Horizontal scaling |
| Storing hierarchical/nested data | Documents support nested structures natively |
| Big Data & real-time analytics | High throughput |
| Content management, catalogs | Flexible document model suits varied content types |


## Q2. State and Explain the features of MongoDB.

### Answer:

1. **Document-Oriented Storage:** Data is stored as BSON documents (similar to JSON), allowing nested fields and arrays.

2. **Schema-less / Flexible Schema:** Each document in a collection can have a different structure — no rigid table schema required.

3. **Indexing:** Supports indexing on any field including nested fields, enabling fast query performance.

4. **Replication (Replica Sets):** Provides high availability through replica sets — a primary node and multiple secondary nodes that sync data.

5. **Sharding (Horizontal Scalability):** Data can be distributed across multiple machines using sharding, enabling horizontal scaling for large datasets.

6. **Aggregation Framework:** Powerful pipeline-based aggregation (like `$group`, `$match`, `$sort`) for data transformation and analysis.

7. **Ad hoc Queries:** Supports queries by field, range queries, and regular expression searches.

8. **GridFS:** A specification for storing and retrieving large files (e.g., images, videos) exceeding the BSON document size limit (16MB).

9. **High Performance:** In-memory processing of data for fast read/write operations.

10. **Load Balancing:** Distributes data across multiple servers automatically.


## Q3. Write a code to connect MongoDB to Python. Also, create a database and a collection in MongoDB.

````python
# Install pymongo if not already installed
# !pip install pymongo

import pymongo

# Step 1: Connect to MongoDB (local instance)
client = pymongo.MongoClient("mongodb://localhost:27017/")

# Step 2: Create a database named 'pwskills_db'
db = client["pwskills_db"]

# Step 3: Create a collection named 'students'
collection = db["students"]

print("Connection successful!")
print("Database: pwskills_db")
print("Collection: students")

# Note: In MongoDB, databases and collections are created lazily —
# they are actually created when the first document is inserted.
````

## Q4. Using the database and the collection created in Q3, write a code to insert one record, and insert many records. Use the find() and find_one() methods to print the inserted records.

````python
# Insert ONE record
one_record = {
    "name": "Chinmay",
    "course": "Data Science",
    "age": 22,
    "grade": "A"
}
result_one = collection.insert_one(one_record)
print("Inserted one record with ID:", result_one.inserted_id)

# Insert MANY records
many_records = [
    {"name": "Alice",   "course": "Machine Learning", "age": 23, "grade": "A+"},
    {"name": "Bob",     "course": "Web Development",  "age": 21, "grade": "B"},
    {"name": "Charlie", "course": "Data Science",     "age": 24, "grade": "A"},
    {"name": "Diana",   "course": "Cloud Computing",  "age": 22, "grade": "A+"}
]
result_many = collection.insert_many(many_records)
print("Inserted many records with IDs:", result_many.inserted_ids)

print("\n--- find_one() output ---")
print(collection.find_one({"name": "Chinmay"}))

print("\n--- find() output (all records) ---")
for doc in collection.find():
    print(doc)
````

## Q5. Explain how you can use the find() method to query the MongoDB database. Write a simple code to demonstrate this.

### Explanation:

The `find()` method is used to **query documents** from a MongoDB collection. It returns a **cursor** object that can be iterated.

**Syntax:**
```python
collection.find(filter, projection)
```
- `filter` — (optional) A dictionary specifying query conditions.
- `projection` — (optional) Specifies which fields to include or exclude.

**Common Query Operators:**
| Operator | Meaning |
|---|---|
| `$eq` | Equal to |
| `$gt` | Greater than |
| `$lt` | Less than |
| `$gte` | Greater than or equal |
| `$lte` | Less than or equal |
| `$in` | Matches any value in a list |
| `$ne` | Not equal to |


````python
# Query 1: Find all documents
print("All students:")
for doc in collection.find():
    print(doc)

# Query 2: Find students with age > 22
print("\nStudents with age > 22:")
for doc in collection.find({"age": {"$gt": 22}}):
    print(doc)

# Query 3: Find students in 'Data Science' course
print("\nData Science students:")
for doc in collection.find({"course": "Data Science"}):
    print(doc)

# Query 4: Find with projection (show only name and grade, hide _id)
print("\nName and Grade only:")
for doc in collection.find({}, {"_id": 0, "name": 1, "grade": 1}):
    print(doc)
````

## Q6. Explain the sort() method. Give an example to demonstrate sorting in MongoDB.

### Explanation:

The `sort()` method is used to **sort the results** of a query in ascending or descending order.

**Syntax:**
```python
collection.find().sort(key, direction)
```
- `key` — The field name to sort by.
- `direction` — `1` for **Ascending**, `-1` for **Descending**.

You can also sort by multiple fields:
```python
collection.find().sort([("field1", 1), ("field2", -1)])
```


````python
import pymongo

# Sort by age in Ascending order
print("Students sorted by Age (Ascending):")
for doc in collection.find({}, {"_id": 0}).sort("age", pymongo.ASCENDING):
    print(doc)

# Sort by age in Descending order
print("\nStudents sorted by Age (Descending):")
for doc in collection.find({}, {"_id": 0}).sort("age", pymongo.DESCENDING):
    print(doc)

# Sort by name in Alphabetical order
print("\nStudents sorted by Name (A-Z):")
for doc in collection.find({}, {"_id": 0}).sort("name", 1):
    print(doc)
````

## Q7. Explain why delete_one(), delete_many(), and drop() is used.

### Explanation:

| Method | Purpose |
|---|---|
| `delete_one(filter)` | Deletes the **first** document matching the filter |
| `delete_many(filter)` | Deletes **all** documents matching the filter |
| `drop()` | **Drops (deletes) the entire collection** including all its documents and indexes |

---

**`delete_one(filter)`:**  
Used when you want to remove a single specific document. If multiple documents match, only the first encountered is deleted.

**`delete_many(filter)`:**  
Used to remove all documents that satisfy a condition. Useful for bulk cleanup operations.

**`drop()`:**  
Completely removes the collection from the database. Unlike `delete_many({})` which removes documents but keeps the collection, `drop()` removes the collection itself.


````python
# delete_one(): Delete the first student named 'Bob'
result = collection.delete_one({"name": "Bob"})
print("delete_one() - Deleted count:", result.deleted_count)

# Verify
print("\nAfter delete_one():")
for doc in collection.find({}, {"_id": 0}):
    print(doc)

# delete_many(): Delete all students in 'Data Science'
result = collection.delete_many({"course": "Data Science"})
print("\ndelete_many() - Deleted count:", result.deleted_count)

# Verify
print("\nAfter delete_many():")
for doc in collection.find({}, {"_id": 0}):
    print(doc)

# drop(): Drop the entire collection
# Uncomment below line to drop the collection
# collection.drop()
# print("\nCollection 'students' has been dropped.")
# print("Collections in DB:", db.list_collection_names())
````
