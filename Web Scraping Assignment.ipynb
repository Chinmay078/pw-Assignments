## Q1. What is Web Scraping? Why is it Used? Give three areas where Web Scraping is used to get data.

Web Scraping is the process of automatically extracting data from websites using software or scripts instead of copying it manually.

**Why is it used?**
- Collect large amounts of data quickly.
- Save time and effort.
- Analyze data for research and business decisions.

**Three application areas:**
1. E-commerce – Price comparison and product monitoring.
2. News & Media – Collecting news articles and headlines.
3. Real Estate – Gathering property prices and listings.

## Q2. What are the different methods used for Web Scraping?

The common methods are:
- Manual Copy-Paste
- HTML Parsing (BeautifulSoup)
- DOM Parsing
- XPath
- Regular Expressions (Regex)
- Browser Automation (Selenium)
- Using Website APIs (when available)

## Q3. What is Beautiful Soup? Why is it used?

Beautiful Soup is a Python library used to parse HTML and XML documents.

**Why is it used?**
- Extract data from web pages.
- Navigate HTML tags easily.
- Search elements using tag names, IDs, and classes.
- Clean and organize scraped data.

## Q4. Why is Flask used in this Web Scraping project?

Flask is a lightweight Python web framework.

It is used to:
- Create a web interface.
- Receive user input.
- Execute the scraping script.
- Display scraped data in the browser.
- Connect frontend and backend.

## Q5. Write the names of AWS services used in this project. Also, explain the use of each service.

Common AWS services used in a web scraping project are:

- **Amazon EC2:** Runs the scraping application.
- **Amazon S3:** Stores scraped data and files.
- **Amazon RDS:** Stores structured data in a relational database.
- **AWS Lambda:** Executes scraping tasks without managing servers.
- **Amazon CloudWatch:** Monitors application logs and performance.
- **IAM (Identity and Access Management):** Controls secure access to AWS resources.
