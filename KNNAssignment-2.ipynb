## Q1. What is the main difference between the Euclidean distance metric and the Manhattan distance metric in KNN? How might this difference affect the performance of a KNN classifier or regressor?

**Answer:**

The primary difference lies in how the distance between two points is calculated.

**Euclidean Distance**

\[
d=\sqrt{\sum_{i=1}^{n}(x_i-y_i)^2}
\]

- Measures the straight-line distance.
- Sensitive to outliers.
- Suitable for continuous data.

**Manhattan Distance**

\[
d=\sum_{i=1}^{n}|x_i-y_i|
\]

- Measures the distance along grid-like paths.
- Less sensitive to outliers.
- Often performs better for high-dimensional or sparse data.

**Effect on Performance**
- Euclidean distance works well when features have similar scales and the relationship is geometric.
- Manhattan distance is preferred when data contains outliers or follows grid-like movement.
- The best metric should be selected using cross-validation.


## Q2. How do you choose the optimal value of K for a KNN classifier or regressor? What techniques can be used to determine the optimal K value?

**Answer:**

The value of K controls the complexity of the model.

- Small K may lead to overfitting.
- Large K may lead to underfitting.

**Techniques to choose K**
1. Cross-validation.
2. Grid Search.
3. Elbow method using validation error.
4. Rule of thumb: K ≈ √N.
5. Select the K giving the highest validation accuracy (classification) or lowest error (regression).


## Q3. How does the choice of distance metric affect the performance of a KNN classifier or regressor? In what situations might you choose one distance metric over the other?

**Answer:**

The distance metric determines which samples are considered nearest neighbors.

- Euclidean distance is suitable for continuous numerical data after feature scaling.
- Manhattan distance performs better for sparse or high-dimensional datasets.
- Minkowski distance generalizes both Euclidean and Manhattan.
- Cosine similarity is useful for text data.

The choice depends on the characteristics of the dataset and should be validated experimentally.


## Q4. What are some common hyperparameters in KNN classifiers and regressors, and how do they affect the performance of the model? How might you go about tuning these hyperparameters to improve model performance?

**Answer:**

Common hyperparameters include:
- Number of neighbors (K)
- Distance metric
- Weight function (uniform or distance)
- Search algorithm (KD-Tree, Ball Tree, Brute Force)

**Effect**
- K controls bias-variance trade-off.
- Distance metric changes neighbor selection.
- Distance weighting gives more importance to closer neighbors.

**Hyperparameter Tuning**
- Grid Search
- Random Search
- Cross-validation


## Q5. How does the size of the training set affect the performance of a KNN classifier or regressor? What techniques can be used to optimize the size of the training set?

**Answer:**

A larger training set generally improves prediction accuracy because more representative neighbors are available.

However:
- Prediction time increases.
- Memory usage increases.

**Optimization Techniques**
- Feature selection
- PCA
- Remove duplicate or noisy samples
- Efficient data structures (KD-Tree/Ball Tree)
- Stratified sampling


## Q6. What are some potential drawbacks of using KNN as a classifier or regressor? How might you overcome these drawbacks to improve the performance of the model?

**Answer:**

**Drawbacks**
- Slow prediction for large datasets.
- Sensitive to feature scaling.
- Sensitive to noise and outliers.
- Suffers from the curse of dimensionality.
- High memory usage.

**Solutions**
- Normalize or standardize features.
- Select an optimal K.
- Remove outliers.
- Apply PCA or feature selection.
- Use KD-Tree or Ball Tree for faster neighbor search.
- Perform cross-validation to choose the best hyperparameters.

These techniques improve both the efficiency and accuracy of KNN models.
