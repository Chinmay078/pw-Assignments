## Q1. What is Elastic Net Regression and how does it differ from other regression techniques?

Elastic Net Regression is a regularized linear regression method that combines **L1 (Lasso)** and **L2 (Ridge)** penalties. It performs feature selection like Lasso while also handling multicollinearity like Ridge.

## Q2. How do you choose the optimal values of the regularization parameters for Elastic Net Regression?

The parameters **alpha** and **l1_ratio** are typically selected using cross-validation (e.g., ElasticNetCV), grid search, or validation datasets to minimize prediction error.

## Q3. What are the advantages and disadvantages of Elastic Net Regression?

**Advantages:** Performs feature selection, handles multicollinearity, reduces overfitting, and works well with high-dimensional data.

**Disadvantages:** Requires tuning multiple parameters, needs feature scaling, and may underfit if regularization is too strong.

## Q4. What are some common use cases for Elastic Net Regression?

Common applications include genomics, finance, healthcare, text mining, marketing analytics, and any dataset with many correlated features.

## Q5. How do you interpret the coefficients in Elastic Net Regression?

The coefficients represent the effect of each feature on the target variable. Regularization shrinks the coefficients, and some may become exactly zero, indicating unimportant features.

## Q6. How do you handle missing values when using Elastic Net Regression?

Missing values should be handled before training using techniques such as mean, median, or mode imputation, KNN imputation, or by removing rows/columns with excessive missing values.

## Q7. How do you use Elastic Net Regression for feature selection?

Train the Elastic Net model and retain only the features whose coefficients are non-zero. Features with coefficients equal to zero can be removed.
