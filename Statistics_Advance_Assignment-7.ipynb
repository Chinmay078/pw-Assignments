## Q1. Python function to calculate the F-value and p-value

```
import numpy as np
from scipy.stats import f

def f_test(sample1, sample2):
    var1 = np.var(sample1, ddof=1)
    var2 = np.var(sample2, ddof=1)
    F = var1 / var2
    dfn = len(sample1) - 1
    dfd = len(sample2) - 1
    p = 2 * min(f.cdf(F, dfn, dfd), 1 - f.cdf(F, dfn, dfd))
    return F, p

a=[12,15,14,10,18]
b=[8,10,12,11,9]
print(f_test(a,b))
```

## Q2. Critical F-value for a two-tailed test

```
from scipy.stats import f
alpha=0.05
dfn=10
dfd=12
critical=f.ppf(1-alpha/2, dfn, dfd)
print("Critical F-value:", critical)
```

## Q3. Generate random samples and perform an F-test

```
import numpy as np
from scipy.stats import f

np.random.seed(42)
x=np.random.normal(50,3,20)
y=np.random.normal(52,4,20)

F=np.var(x,ddof=1)/np.var(y,ddof=1)
df1=len(x)-1
df2=len(y)-1
p=2*min(f.cdf(F,df1,df2),1-f.cdf(F,df1,df2))
print(F,df1,df2,p)
```

## Q6. Function to calculate mean and variance of F-distribution

```
def f_distribution_stats(df1, df2):
    mean = df2/(df2-2) if df2>2 else None
    variance = (2*(df2**2)*(df1+df2-2))/(df1*((df2-2)**2)*(df2-4)) if df2>4 else None
    return mean, variance

print(f_distribution_stats(10,12))
```

## Q8. Restaurant waiting times

```
import numpy as np
from scipy.stats import f

A=[24,25,28,23,22,20,27]
B=[31,33,35,30,32,36]
F=np.var(A,ddof=1)/np.var(B,ddof=1)
p=2*min(f.cdf(F,6,5),1-f.cdf(F,6,5))
print(F,p)
```
