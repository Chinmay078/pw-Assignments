# Q1. Purpose of Forward Propagation

Forward propagation is the process of passing input data through a neural network to generate predictions. Inputs are multiplied by weights, biases are added, activation functions are applied, and the output is passed to the next layer until the final prediction is obtained.

# Q2. Mathematical implementation

For a single-layer network:

z = Wx + b

a = f(z)

where W is the weight matrix, x is the input vector, b is the bias, and f is the activation function.

# Q3. Activation Functions

Activation functions introduce non-linearity so neural networks can learn complex relationships. Common functions include ReLU, Sigmoid, Tanh, and Softmax.

# Q4. Weights and Biases

Weights determine the importance of input features, while biases shift the activation values. Both are updated during training to minimize prediction error.

# Q5. Softmax Function

Softmax converts output values into probabilities whose sum equals 1. It is mainly used in the output layer for multi-class classification.

# Q6. Purpose of Backward Propagation

Backward propagation computes gradients of the loss function with respect to weights and biases. These gradients are used to update parameters and improve model performance.

# Q7. Mathematical calculation

Error is propagated backward using the chain rule.

Weight update:
W = W − η(dL/dW)

Bias update:
b = b − η(dL/db)

where η is the learning rate.

# Q8. Chain Rule

The chain rule breaks complex derivatives into smaller parts, allowing gradients to be efficiently computed for every layer in the network.

# Q9. Challenges

Common challenges include vanishing gradients, exploding gradients, overfitting, poor weight initialization, and improper learning rates. Solutions include ReLU, Batch Normalization, Adam optimizer, gradient clipping, dropout, and better initialization.
