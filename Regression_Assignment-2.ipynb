## Q1. R-squared in Linear Regression Models

R-squared (R²) measures the proportion of variation in the dependent variable explained by the independent variables. R² = 1 − (SSR/SST). A higher R² indicates a better fit.

## Q2. Adjusted R-squared

Adjusted R² accounts for the number of predictors and prevents misleading improvements when unnecessary variables are added. Unlike R², it can decrease if a useless feature is included.

## Q3. When is Adjusted R-squared appropriate?

Use Adjusted R² when comparing regression models with different numbers of independent variables.

## Q4. RMSE, MSE and MAE

MSE is the average squared error, RMSE is the square root of MSE, and MAE is the average absolute error. Lower values indicate better predictions.

## Q5. Advantages and disadvantages

MSE and RMSE penalize large errors heavily, while MAE is more robust to outliers and easier to interpret.

## Q6. Lasso Regularization

Lasso (L1) shrinks coefficients and can reduce some to zero, performing feature selection. Ridge (L2) only shrinks coefficients.

## Q7. Regularized models and overfitting

Regularization reduces model complexity by penalizing large coefficients, improving generalization and reducing overfitting.

## Q8. Limitations

Regularized models assume linearity, require parameter tuning, and may underfit if regularization is too strong.

## Q10. Ridge vs Lasso

Choose Ridge when all predictors are useful. Choose Lasso when feature selection is needed. Use cross-validation to tune the regularization parameter.
