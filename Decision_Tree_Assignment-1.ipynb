## Q1. Describe the Decision Tree classifier algorithm and how it works to make predictions.

Decision Tree is a supervised learning algorithm used for classification. It splits data into smaller subsets using feature-based decision rules. Internal nodes represent decisions, branches represent outcomes, and leaf nodes represent the predicted class.

## Q2. Mathematical intuition behind Decision Trees.

The algorithm computes impurity (Gini Index or Entropy), selects the split with the highest Information Gain, and recursively repeats the process until stopping criteria are met.

## Q3. Decision Tree for binary classification.

It predicts one of two classes. Example: Spam vs. Not Spam email classification.

## Q4. Geometric intuition.

Decision Trees partition the feature space into rectangular regions using axis-aligned splits. Each region corresponds to a predicted class.

## Q5. Confusion Matrix.

A confusion matrix contains TP, TN, FP, and FN. It is used to evaluate the performance of classification models.

## Q6. Precision, Recall and F1-score example.

Example: TP=90, FP=10, FN=20, TN=80.
Precision=0.90, Recall=0.82, F1≈0.86.

## Q7. Choosing an evaluation metric.

Accuracy is suitable for balanced datasets, while Precision, Recall, F1-score, and ROC-AUC are preferred for imbalanced datasets.

## Q8. When Precision is most important.

Spam email detection is an example where false positives should be minimized.

## Q9. When Recall is most important.

Disease diagnosis is an example where missing positive cases should be avoided.
