## Q1. What is boosting in machine learning?

Boosting is an ensemble learning technique that combines multiple weak learners to create a strong learner. Weak learners (usually decision stumps or shallow decision trees) are trained sequentially. Each new model focuses more on the mistakes made by the previous models. The final prediction is obtained by combining the weighted predictions of all weak learners.

**Example:** AdaBoost combines many decision stumps to achieve high classification accuracy.

## Q2. Advantages and limitations of boosting techniques

**Advantages**
- Improves prediction accuracy.
- Reduces bias and variance.
- Works well with weak learners.
- Can capture complex relationships.
- Widely used in competitions.

**Limitations**
- Sensitive to noisy data and outliers.
- Training is slower because models are built sequentially.
- Can overfit if too many estimators are used.
- Requires careful hyperparameter tuning.

## Q3. Explain how boosting works.

1. Assign equal weights to all training samples.
2. Train the first weak learner.
3. Identify incorrectly classified samples.
4. Increase their weights and decrease weights of correctly classified samples.
5. Train the next learner on the updated weights.
6. Repeat for several iterations.
7. Combine all learners using weighted voting (classification) or weighted averaging (regression).

## Q6. How do boosting algorithms combine weak learners?

Each weak learner is trained sequentially to correct previous errors. Better-performing learners receive higher weights. The final prediction is the weighted combination of all weak learners, resulting in a strong model with higher accuracy.

## Q7. Explain AdaBoost and its working.

AdaBoost (Adaptive Boosting) is an ensemble algorithm that repeatedly trains weak learners while updating sample weights.

Working:
1. Initialize equal sample weights.
2. Train a weak learner.
3. Compute its error.
4. Calculate learner weight:
   α = 1/2 × ln((1-error)/error)
5. Increase weights of misclassified samples.
6. Normalize weights.
7. Repeat for the desired number of estimators.
8. Final prediction is the weighted vote of all learners.

## Q8. Loss function used in AdaBoost

AdaBoost minimizes the **Exponential Loss**:

L = exp(-yF(x))

where y is the true label and F(x) is the ensemble prediction. This loss heavily penalizes misclassified samples.

## Q9. How does AdaBoost update weights of misclassified samples?

After each iteration:
- Misclassified samples receive higher weights.
- Correctly classified samples receive lower weights.
- Weights are normalized before the next iteration.

This forces the next weak learner to focus on difficult examples.

## Q10. Effect of increasing the number of estimators

Increasing the number of estimators generally improves performance up to a point because more weak learners correct previous mistakes. However:
- Training time increases.
- Very large values may lead to overfitting (especially with noisy data).
- A suitable value is selected using validation or cross-validation.
