## Q1. What is Flask Framework? What are the advantages of Flask Framework?

**Flask** is a lightweight Python web framework used to build web applications and APIs.

**Advantages**
- Lightweight and simple
- Easy to learn
- Flexible
- Built-in development server
- REST API support
- Large community support


## Q2. Create a simple Flask application to display 'Hello World!!'.

```
from flask import Flask

app = Flask(__name__)

@app.route('/')
def home():
    return "Hello World!!"

if __name__ == '__main__':
    app.run(debug=True)
```

## Q3. What is App Routing in Flask? Why do we use app routes?

App routing maps a URL to a Python function using the `@app.route()` decorator.

**Why use routes?**
- Connect URLs to functions
- Organize web pages
- Handle user requests
- Build REST APIs


## Q4. Create `/welcome` and `/` routes.

```
from flask import Flask

app = Flask(__name__)

@app.route('/')
def company():
    return '''
    <h2>Company Name: ABC Corporation</h2>
    <h3>Location: India</h3>
    <h3>Contact Detail: 999-999-9999</h3>
    '''

@app.route('/welcome')
def welcome():
    return "Welcome to ABC Corporation"

if __name__ == '__main__':
    app.run(debug=True)
```

## Q5. Demonstrate `url_for()` in Flask.

```
from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def home():
    return "Home Page"

with app.test_request_context():
    print(url_for('home'))
```
