## Q1. What is Lasso Regression, and how does it differ from other regression techniques?

Lasso (Least Absolute Shrinkage and Selection Operator) is a linear regression technique that uses **L1 regularization**. It adds the absolute values of the coefficients to the loss function, shrinking some coefficients exactly to zero. Unlike ordinary linear regression, it performs both regularization and feature selection.

## Q2. What is the main advantage of using Lasso Regression in feature selection?

The main advantage is that Lasso automatically removes less important features by setting their coefficients to zero, resulting in a simpler and more interpretable model.

## Q3. How do you interpret the coefficients of a Lasso Regression model?

The coefficients indicate the effect of each predictor on the target variable. Coefficients that become zero are excluded from the model, while larger absolute coefficients have greater influence.

## Q4. What are the tuning parameters that can be adjusted in Lasso Regression, and how do they affect the model's performance?

The primary tuning parameter is **lambda (alpha)**. Increasing alpha increases regularization, shrinking more coefficients toward zero. Decreasing alpha reduces regularization and makes the model closer to ordinary linear regression.

## Q5. Can Lasso Regression be used for non-linear regression problems? If yes, how?

Yes. Lasso can be applied to non-linear problems after creating polynomial or other transformed features. The transformed features are then used in a Lasso regression model.

## Q6. What is the difference between Ridge Regression and Lasso Regression?

- **Lasso (L1):** Can set coefficients exactly to zero and performs feature selection.
- **Ridge (L2):** Shrinks coefficients but usually does not make them zero. It is better when all features are useful.

## Q7. Can Lasso Regression handle multicollinearity in the input features? If yes, how?

Yes. Lasso helps reduce multicollinearity by selecting one or a few important correlated features while shrinking the others to zero.

## Q8. How do you choose the optimal value of the regularization parameter (lambda) in Lasso Regression?

The optimal value is usually selected using **cross-validation** (e.g., LassoCV), Grid Search, or validation data. The value that minimizes prediction error is chosen.
