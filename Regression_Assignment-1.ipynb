## Q1. Difference between simple and multiple linear regression

**Simple Linear Regression** uses one independent variable to predict one dependent variable.

Example: Predicting salary based on years of experience.

**Multiple Linear Regression** uses two or more independent variables.

Example: Predicting house prices using area, number of bedrooms, and location.

## Q2. Assumptions of linear regression

The main assumptions are:
- Linearity
- Independence of observations
- Homoscedasticity (constant variance)
- Normally distributed residuals
- No multicollinearity

These can be checked using residual plots, Q-Q plots, VIF, and statistical tests.

## Q3. Interpret slope and intercept

The **intercept** is the predicted value of the dependent variable when all predictors are zero.

The **slope** indicates how much the dependent variable changes for a one-unit increase in the predictor.

Example: If Salary = 20000 + 5000 × Experience, each additional year of experience increases salary by ₹5,000.

## Q4. Gradient Descent

Gradient Descent is an optimization algorithm used to minimize the loss function by updating model parameters in the direction of the negative gradient.

It is widely used to train machine learning models such as linear regression and neural networks.

## Q5. Multiple Linear Regression Model

Multiple Linear Regression models the relationship between one dependent variable and multiple independent variables.

Unlike simple linear regression, it considers the combined effect of several predictors.

## Q6. Multicollinearity

Multicollinearity occurs when independent variables are highly correlated.

Detection methods:
- Correlation matrix
- Variance Inflation Factor (VIF)

Solutions:
- Remove correlated features
- Apply PCA
- Use Ridge/Lasso regression.

## Q7. Polynomial Regression

Polynomial Regression models nonlinear relationships by adding polynomial terms (x², x³, etc.).

Unlike linear regression, it can fit curved relationships while remaining linear in the coefficients.

## Q8. Advantages and disadvantages of Polynomial Regression

**Advantages**
- Captures nonlinear relationships.
- Better fit for curved data.

**Disadvantages**
- Can overfit.
- Sensitive to polynomial degree.
- Harder to interpret.

Use Polynomial Regression when the relationship between variables is nonlinear.
